```python
import numpy as np
import matplotlib.pyplot as plt
```

### Parameters

```python
dt      = 0.01
k_on    = 0.1
k_off   = 0.1
a       = 3
p       = 10
c0      = 1
n_steps = 1200
l_star  = (k_on*p - k_off*c0) / (2*k_on + a*k_off) 
```

```python
# Initialize the lenth
l1 = np.zeros(n_steps)
l2 = np.zeros(n_steps)
l1 = np.zeros(n_steps)
l2 = np.zeros(n_steps)
ls = np.full(n_steps, l_star)

# Initial condition
#l1[0] = ls

for i in range(1, n_steps):
    l1[i] = l1[i-1] + k_on*(p - l1[i-1] - l2[i-1])*dt - k_off*(a*l1[i-1] + c0)*dt
    l2[i] = l2[i-1] + k_on*(p - l1[i-1] - l2[i-1])*dt - k_off*(a*l2[i-1] + c0)*dt

# Define array for measuring time
times = np.arange(n_steps) * dt

# plot the results
plt.plot(times, ls, label='$L^*$')
plt.plot(times, l1, label='L1')
plt.plot(times, l2, label='L2')
plt.legend();
```

```python
p_cut = p - l_star
l_star_cut  = (k_on*p_cut - k_off*c0) / (2*k_on + a*k_off)

# Initialize the lenth
l1 = np.zeros(n_steps)
l2 = np.zeros(n_steps)
l1 = np.zeros(n_steps)
l2 = np.zeros(n_steps)
ls = np.full(n_steps, l_star)
ls_cut = np.full(n_steps, l_star_cut)

# Initial condition
l1[0] = l_star

for i in range(1, n_steps):
    l1[i] = l1[i-1] + k_on*(p_cut - l1[i-1] - l2[i-1])*dt - k_off*(a*l1[i-1] + c0)*dt
    l2[i] = l2[i-1] + k_on*(p_cut - l1[i-1] - l2[i-1])*dt - k_off*(a*l2[i-1] + c0)*dt

# Define array for measuring time
times = np.arange(n_steps) * dt

# plot the results
plt.plot(times, ls, linestyle = ':', label='$L^*$')
plt.plot(times, ls_cut, linestyle = ':', label='$L^*$ cut')

plt.plot(times, l1, label='L1')
plt.plot(times, l2, label='L2')

plt.legend();
```
