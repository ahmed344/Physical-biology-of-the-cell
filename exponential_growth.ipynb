### Exponential growth
$$ N(t) = N(t - \Delta t) -  (k \Delta t) N(t - \Delta t)$$

$$ \frac{dN}{dt} = k N$$

$$ N(t) = N_0 \exp(k t)$$

$$ t_{doubling} = \frac{\ln{2}}{k}$$

$$ k = \frac{\ln{2}}{t_{doubling}}$$

```python
# specify initial condition and parameters
N0 = 1  # bacterium

# Specify step size
dt = 0.1 # minutes
total_time = 120 # minutes

# Specify the growth rate
k = np.log(2) /30 # min^-1

# Compute the number of steps
num_steps = int(total_time / dt)

# Create list Nt
Nt = np.zeros(num_steps)
Nt[0] = N0
```

```python
# Numerically integrate by looping in time
for t in range(1, num_steps):
    Nt[t] = Nt[t-1] + k*dt*Nt[t-1]
```

```python
# make array to represent time
times = np.arange(num_steps) * dt
```

```python
# Plot th results
plt.plot(times, Nt)
plt.xlabel('time [min]')
plt.ylabel('N');
```

```python
# Compute the known solution
soln = N0 * np.exp(k * times)
```

```python
# Plot the results
plt.plot(times, Nt, label='Numerical', alpha=0.5)
plt.plot(times, soln, label='Analytical', alpha=0.5)
plt.xlabel('time [min]')
plt.ylabel('N')
plt.legend();
```
